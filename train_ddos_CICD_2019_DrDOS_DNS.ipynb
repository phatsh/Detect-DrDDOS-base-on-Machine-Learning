```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler  # Đảm bảo nhập StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
from imblearn.over_sampling import SMOTE
import joblib
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
%matplotlib inline
```

```python
import pandas as pd
import numpy as np

# Định nghĩa chunksize (số dòng mỗi chunk)
CHUNKSIZE = 10000  # Điều chỉnh tùy theo RAM, ví dụ: 10,000 dòng mỗi chunk

# Chỉ định dtype cho cột 85 (SimillarHTTP) để tránh DtypeWarning
dtypes = {85: str}

# Đọc dữ liệu theo chunks
chunk_iter = pd.read_csv(
    '/content/drive/MyDrive/Datasett/CSV-01-12/01-12/DrDoS_DNS.csv',
    dtype=dtypes,
    chunksize=CHUNKSIZE,
    low_memory=False
)

# Khởi tạo danh sách để lưu thông tin nhãn
label_counts = pd.Series(dtype=int)

# Xử lý từng chunk
first_chunk = True
for chunk in chunk_iter:
    # Loại bỏ khoảng trắng trong tên cột
    chunk.columns = chunk.columns.str.strip()

    # Hiển thị thông tin chunk đầu tiên
    if first_chunk:
        print("Thông tin chunk đầu tiên:")
        print(chunk.info())
        print("\nDanh sách các cột trong dataset:")
        print(chunk.columns.tolist())
        first_chunk = False

    # Kiểm tra cột 'Label' hoặc 'label'
    label_col = 'Label' if 'Label' in chunk.columns else 'label' if 'label' in chunk.columns else None
    if label_col:
        # Cộng dồn phân bố nhãn
        chunk_counts = chunk[label_col].value_counts()
        label_counts = label_counts.add(chunk_counts, fill_value=0)
    else:
        print("\nLỗi: Cột 'Label' hoặc 'label' không tồn tại trong chunk!")
        similar_columns = [col for col in chunk.columns if 'label' in col.lower()]
        print("Các cột có tên gần giống:", similar_columns)
        if similar_columns:
            print(f"Gợi ý: Thử sử dụng cột '{similar_columns[0]}'.")
            chunk_counts = chunk[similar_columns[0]].value_counts()
            label_counts = label_counts.add(chunk_counts, fill_value=0)
        else:
            print("Không tìm thấy cột nào liên quan đến 'label'. Vui lòng kiểm tra file CSV.")
        break  # Thoát nếu không tìm thấy cột nhãn

# In phân bố nhãn tổng thể
print("\nPhân bố nhãn tổng thể:")
print(label_counts)
```

```python
# Khởi tạo danh sách để lưu dữ liệu đã xử lý
processed_chunks = []
max_chunks = 5  # Giới hạn số chunk để tránh tràn RAM

# Đọc lại chunk_iter để xử lý
chunk_iter = pd.read_csv(
    '/content/drive/MyDrive/Datasett/CSV-01-12/01-12/DrDoS_DNS.csv',
    dtype=dtypes,
    chunksize=CHUNKSIZE,
    low_memory=False
)

# Xử lý từng chunk
for i, chunk in enumerate(chunk_iter):
    if i >= max_chunks:
        break

    # Loại bỏ khoảng trắng trong tên cột
    chunk.columns = chunk.columns.str.strip()

    # Loại bỏ các cột không cần thiết
    columns_to_drop = ['Unnamed: 0', 'Flow ID', 'Source IP', 'Destination IP', 'Timestamp', 'SimillarHTTP']
    chunk = chunk.drop(columns=columns_to_drop, errors='ignore')

    # Kiểm tra cột nhãn
    if not label_col:
        label_col = 'Label' if 'Label' in chunk.columns else 'label' if 'label' in chunk.columns else None
        if not label_col:
            similar_columns = [col for col in chunk.columns if 'label' in col.lower()]
            label_col = similar_columns[0] if similar_columns else None
            if not label_col:
                print("Lỗi: Không tìm thấy cột nhãn!")
                break

    # Chuyển đổi nhãn thành nhị phân
    chunk[label_col] = chunk[label_col].apply(lambda x: 1 if 'DrDoS' in str(x) else 0)

    # Xử lý giá trị thiếu và vô cực
    chunk = chunk.replace([np.inf, -np.inf], np.nan).fillna(0)

    # Xử lý giá trị âm
    numeric_columns = chunk.select_dtypes(include=[np.number]).columns
    chunk[numeric_columns] = chunk[numeric_columns].clip(lower=0)

    # Lưu chunk đã xử lý
    processed_chunks.append(chunk)

# Hợp nhất các chunk
data = pd.concat(processed_chunks, ignore_index=True)
print("Kích thước dữ liệu sau hợp nhất:", data.shape)

# Lưu dữ liệu đã xử lý
data.to_csv('/content/drive/MyDrive/Datasett/CSV-01-12/01-12/DrDoS_DNS_processed.csv', index=False)
print("Dữ liệu đã xử lý được lưu vào 'DrDoS_DNS_processed.csv'")
```

```python
# Đọc dữ liệu đã xử lý
data = pd.read_csv('/content/drive/MyDrive/Datasett/CSV-01-12/01-12/DrDoS_DNS_processed.csv', low_memory=False)

# Tách features và label
X = data.drop(label_col, axis=1)
y = data[label_col]

# Chuẩn hóa dữ liệu
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Cân bằng dữ liệu bằng SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_scaled, y)
print("Kích thước dữ liệu sau SMOTE:", X_resampled.shape)

# Chia dữ liệu thành tập huấn luyện và kiểm tra
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.3, random_state=42)
print("Kích thước tập huấn luyện:", X_train.shape)
print("Kích thước tập kiểm tra:", X_test.shape)
```

```python
# Huấn luyện mô hình
rf_model = RandomForestClassifier(n_estimators=350, random_state=42, n_jobs=-1)
rf_model.fit(X_train, y_train)
print("Huấn luyện mô hình hoàn tất!")
```

```python
# Dự đoán trên tập kiểm tra
y_pred = rf_model.predict(X_test)

# In ma trận nhầm lẫn và báo cáo phân loại
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
# Lấy mẫu ví dụ
sample = data.drop(label_col, axis=1).iloc[0:1]
sample_scaled = scaler.transform(sample)
prediction = rf_model.predict(sample_scaled)
print("\nDự đoán cho mẫu ví dụ:", "DDoS" if prediction[0] == 1 else "Bình thường")
```
